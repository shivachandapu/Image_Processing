```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
from math import*

# Read the image and display it
img = cv.imread('/home/shiva/Desktop/4_MTCS/SE_2/Shiva Krishna - Noisy_Smithsonian_Castle.jpg',0)
plt.rcParams['figure.figsize'] = [10,10]
plt.imshow(img, cmap = 'gray')
```

```python
# Finding the magnitude spectrum of the given image
fourier_transform = np.fft.fft2(img)
center_shift = np.fft.fftshift(fourier_transform)

# Magnitude spectrum
spectrum = np.log(np.abs(center_shift))

# Printing the magnitude spectrum
plt.show()
plt.imshow(spectrum,cmap = 'gray')
```

```python
# Dimensions of the given image
height, width = img.shape
rows, cols = height//2, width//2

# Blocking the noise
center_shift[:rows-20, cols-2:cols+2] = 0
center_shift[rows+20:, cols-2:cols+2] = 0

# New magnitude spectrum
no_noise_spectrum = np.log(1 + np.abs(center_shift))

# Displaying the new magnitude spectrum
plt.show()
plt.imshow(no_noise_spectrum,cmap='gray')
```

```python
# Finding the inverse FFT
inv_shift = np.fft.ifftshift(center_shift)
denoised_image = np.real(np.fft.ifft2(inv_shift))

# Displaying the denoised image
plt.show()
plt.imshow(denoised_image, cmap='gray')
```

```python
# Finding the noise and then displaying it
noise = img - denoised_image
plt.show()
plt.imshow(noise, cmap = 'gray')
```
